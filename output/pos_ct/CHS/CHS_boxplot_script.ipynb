```python
import seaborn as sns 
import pandas as pd
```

```python
df = pd.read_pickle("CHS_param_file.pkl")
print(df.head(2))
```

```python
df_plot = sns.boxplot(data=df, x="state_names", y="starting_probabilities", hue="pop")
print(df_plot)
```

```python
df_plot_1 = sns.boxenplot(data=df, x="state_names", y="emissions", hue="pop")
print(df_plot_1)
```

```python
new_df_plot = sns.boxplot(data=new_df, x='state_names', y='transition_to_Human', hue='pop')

print(new_df_plot)
```
